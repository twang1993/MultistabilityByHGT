```python
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
from scipy.integrate import solve_ivp
import seaborn as sns
```

```python
global mu10,mu20,mu1,mu2,gamma,D,eta,kappa
```

```python
def TwoSpeciesHGT(t,y):
    s1=y[0]
    s2=y[1]
    p1=y[2]
    p2=y[3]
    lambda1=mu1/mu10-1
    lambda2=mu2/mu20-1
    dydt=[mu1/s1*(s1+lambda2*p1)*s1*(1-s1-gamma*s2)-D*s1,
        mu2/s2*(s2+lambda1*p2)*s2*(1-gamma*s1-s2)-D*s2,
        mu1*(1+lambda2)*p1*(1-s1-gamma*s2)+eta*(s2+p1)*(s1-p1)-kappa*p1-D*p1,
        mu2*(1+lambda1)*p2*(1-gamma*s1-s2)+eta*(s1+p2)*(s2-p2)-kappa*p2-D*p2]
    return dydt
```

```python
gamma=1.1
D=0.2
kappa=0.005
mu1=0.4
mu2=0.6
mu10=0.5
mu20=0.5
timespan=np.arange(0,400.1,0.1)
num=100;
```

```python
eta=0
plt.figure(figsize=(2,2))
for i in range(num):
    initialHGT=[np.random.rand(),np.random.rand(),0,0]
    sol=solve_ivp(TwoSpeciesHGT,[0,max(timespan)],initialHGT,method='RK45',t_eval=timespan)
    y1=np.transpose(sol.y)
    S1Abun=y1[:,0]
    S2Abun=y1[:,1]
    plt.plot(timespan,S1Abun,'-',linewidth=2,color=np.divide([186,222,245],256))
plt.xlabel('time',fontsize=16)
plt.ylabel('$s_1$',fontsize=16)
plt.ylim(0,1)
plt.tick_params(left=False,bottom=False)
ax=plt.gca()
ax.spines[['top', 'right']].set_visible(False)
plt.savefig('Trajectories_D02_eta0.pdf')
plt.show()
```
