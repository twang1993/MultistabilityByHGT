```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
NumStab=pd.read_excel('NumStab01.xlsx').values
```

```python
lambdaa=np.arange(0.1,0.6,0.1)
plt.figure(figsize=(3,3))
plt.errorbar(lambdaa,np.mean(NumStab,0),yerr=np.std(NumStab,0),marker='o',fillstyle='none',markersize=15,linewidth=2)
plt.xlabel('fitness effect range',fontsize=20)
plt.ylabel('regional diversity',fontsize=20)
plt.tick_params(direction='in',top='true',right='true')
plt.ylim(0,6)
plt.savefig('ReDraw_1.pdf')
plt.show()
```
