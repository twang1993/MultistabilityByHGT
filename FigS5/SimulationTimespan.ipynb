```python
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
from scipy.integrate import solve_ivp
from math import dist
import pandas as pd
```

```python
global NumSpecies,mu,lambdaa,gamma,delta,D,eta,kappa,Ip
```

```python
def MultiSpecies(t,y):
    dydt=np.zeros(NumSpecies*(1+NumSpecies))
    for i in range(NumSpecies):
        if y[i]==0:
            dydt[i]=Ip[i]
            for j in range(NumSpecies):
                dydt[NumSpecies+i*NumSpecies+j]=0
        else:
            ss=1
            sumy_neg=0
            sumy_pos=0
            for j in range(NumSpecies):
                if i!=j:
                    ss=ss*(1+lambdaa[j]*y[NumSpecies+i*NumSpecies+j]/y[i])
                if gamma[j,i]<0:
                    sumy_neg=sumy_neg+gamma[j,i]*y[j]
                if gamma[j,i]>0:
                    sumy_pos=sumy_pos+gamma[j,i]*y[j]
            dydt[i]=mu[i]*y[i]*ss*(1+sumy_neg+delta[i]*sumy_pos/(1+sumy_pos))-D*y[i]+Ip[i]
            for j in range(NumSpecies):
                if j==i:
                    dydt[NumSpecies+i*NumSpecies+j]=dydt[i]
                else:
                    ss=1
                    for k in range(NumSpecies):
                        if k!=i and k!=j:
                            ss=ss*(1+lambdaa[k]*y[NumSpecies+i*NumSpecies+k]/y[i])
                    donor=0
                    for k in range(NumSpecies):
                        donor=donor+y[NumSpecies+k*NumSpecies+j]
                    dydt[NumSpecies+i*NumSpecies+j]=mu[i]*y[NumSpecies+i*NumSpecies+j]*ss*(1+lambdaa[j])*(1+sumy_neg+delta[i]*sumy_pos/(1+sumy_pos))+eta*(y[i]-y[NumSpecies+i*NumSpecies+j])*donor-(kappa+D)*y[NumSpecies+i*NumSpecies+j]
    return dydt
```

```python
NumSpecies=5
mu0=0.5
eta=0.2
repeat=200
Ip=0*np.ones(NumSpecies)
neg_fraq=1
pos_fraq=0
delta=0.2*np.ones(NumSpecies)
kappa=0.005
D=0.2
```

```python
timespan=np.arange(0,2000.1,.1)
nn=10
Abun500=np.zeros((nn,repeat,NumSpecies))
Abun1000=np.zeros((nn,repeat,NumSpecies))
Abun1500=np.zeros((nn,repeat,NumSpecies))
Abun2000=np.zeros((nn,repeat,NumSpecies))
for zxc in range(nn):
    lambdaa=-0.4+0.8*np.random.rand(NumSpecies)
    mu=np.multiply(mu0,1+lambdaa)
    gamma=np.zeros((NumSpecies,NumSpecies))
    for i in range(NumSpecies):
        #for j in np.arange(i,NumSpecies,1):
        for j in range(NumSpecies):
            if i==j:
                gamma[i,j]=-1
            else:
                temp=np.random.rand()
                if temp<neg_fraq:
                    gamma[i,j]=-1.05#(1+0.5*np.random.rand())
                    #gamma[j,i]=-(1+0.5*np.random.rand())
                if temp>1-pos_fraq:
                    gamma[i,j]=0.5
                    #gamma[j,i]=0.5
    for hjk in range(repeat):
        print(zxc*repeat+hjk)
        initial=np.zeros(NumSpecies**2+NumSpecies)
        for j in range(NumSpecies):
            initial[j]=np.random.rand()#*2/NumSpecies
            initial[NumSpecies+j*NumSpecies+j]=initial[j]          
        sol=solve_ivp(MultiSpecies,[0,max(timespan)],initial,method='RK45',t_eval=timespan)
        y=np.transpose(sol.y)
        Abun1000[zxc,hjk,:]=y[10000,range(NumSpecies)]  
        Abun1500[zxc,hjk,:]=y[15000,range(NumSpecies)]  
        Abun2000[zxc,hjk,:]=y[20000,range(NumSpecies)]    
```

```python
NumStab=np.zeros((nn,3))
MetaDiver=np.zeros((nn,3))
thresh=0.05

for zxc in range(nn):
    Index=np.zeros(repeat)
    Index[0]=1
    for j in np.arange(1,repeat,1):
        tempj=Abun1000[zxc,j,:]
        pin=0
        for k in range(j):
            tempk=Abun1000[zxc,k,:]
            tttt=tempj-tempk
            dis=np.sqrt(sum(np.multiply(tttt,tttt)))
            if dis<thresh:
                Index[j]=Index[k]
                pin=1
        if pin==0:
            Index[j]=max(Index)+1

    NumStab[zxc,0]=max(Index)
    AbunTot=np.zeros(int(np.max(Index)))
    for rgn in range(int(max(Index))):
        AbunTot[rgn]=np.sum(Index==rgn+1)
    AbunTot=np.divide(AbunTot,sum(AbunTot))
    MetaDiver[zxc,0]=np.exp(np.sum(-np.multiply(AbunTot,np.log(AbunTot))))

for zxc in range(nn):
    Index=np.zeros(repeat)
    Index[0]=1
    for j in np.arange(1,repeat,1):
        tempj=Abun1500[zxc,j,:]
        pin=0
        for k in range(j):
            tempk=Abun1500[zxc,k,:]
            tttt=tempj-tempk
            dis=np.sqrt(sum(np.multiply(tttt,tttt)))
            if dis<thresh:
                Index[j]=Index[k]
                pin=1
        if pin==0:
            Index[j]=max(Index)+1

    NumStab[zxc,1]=max(Index)
    AbunTot=np.zeros(int(np.max(Index)))
    for rgn in range(int(max(Index))):
        AbunTot[rgn]=np.sum(Index==rgn+1)
    AbunTot=np.divide(AbunTot,sum(AbunTot))
    MetaDiver[zxc,1]=np.exp(np.sum(-np.multiply(AbunTot,np.log(AbunTot))))

for zxc in range(nn):
    Index=np.zeros(repeat)
    Index[0]=1
    for j in np.arange(1,repeat,1):
        tempj=Abun2000[zxc,j,:]
        pin=0
        for k in range(j):
            tempk=Abun2000[zxc,k,:]
            tttt=tempj-tempk
            dis=np.sqrt(sum(np.multiply(tttt,tttt)))
            if dis<thresh:
                Index[j]=Index[k]
                pin=1
        if pin==0:
            Index[j]=max(Index)+1

    NumStab[zxc,2]=max(Index)
    AbunTot=np.zeros(int(np.max(Index)))
    for rgn in range(int(max(Index))):
        AbunTot[rgn]=np.sum(Index==rgn+1)
    AbunTot=np.divide(AbunTot,sum(AbunTot))
    MetaDiver[zxc,2]=np.exp(np.sum(-np.multiply(AbunTot,np.log(AbunTot))))
```

```python
plt.figure(figsize=(3,3))
plt.plot([1000,1500,2000],np.transpose(MetaDiver),'o-',markersize=10,fillstyle='none')
plt.xlabel('Simulation Timespan',fontsize=20)
plt.ylabel('multistability',fontsize=20)
plt.tick_params(direction='in',top='true',right='true')
plt.savefig('SimulationTimespan.pdf')
plt.show()

df=pd.DataFrame(MetaDiver)
df.to_csv("SimulationTimespan.csv")
```
