```python
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
from scipy.integrate import solve_ivp
import seaborn as sns
```

```python
global mu10,mu20,mu1,mu2,gamma,D,eta,kappa
```

```python
def TwoSpeciesHGT(t,y):
    s1=y[0]
    s2=y[1]
    p1=y[2]
    p2=y[3]
    lambda1=mu1/mu10-1
    lambda2=mu2/mu20-1
    dydt=[mu1/s1*(s1+lambda2*p1)*s1*(1-s1-gamma*s2)-D*s1,
        mu2/s2*(s2+lambda1*p2)*s2*(1-gamma*s1-s2)-D*s2,
        mu1*(1+lambda2)*p1*(1-s1-gamma*s2)+eta*(s2+p1)*(s1-p1)-kappa*p1-D*p1,
        mu2*(1+lambda1)*p2*(1-gamma*s1-s2)+eta*(s1+p2)*(s2-p2)-kappa*p2-D*p2]
    return dydt
```

```python
gammas=[0.8,0.9,1,1.1,1.2]
D=0.2
etas=np.arange(0,0.45,0.1)
kappa=0.005

rep=500
mu10=0.5
mu20=0.5
num=100
timespan=np.arange(0,500.1,0.1)

feasi=np.zeros((len(etas),len(gammas)))
thresh=0.02

for wer in range(len(gammas)):
    gamma=gammas[wer]
    for sdf in range(len(etas)):
        eta=etas[sdf]
        S1AbunHGT=np.zeros((rep,num))
        S2AbunHGT=np.zeros((rep,num))
        Indd=np.zeros(rep)
        for i in range(rep):
            print(sdf*rep+i)
            mu1=np.random.rand()
            mu2=np.random.rand()
            for k in range(num):
                temp1=np.random.rand()
                temp2=np.random.rand()
                initial=[temp1,temp2]
                initialHGT=[temp1,temp2,0,0]
                sol=solve_ivp(TwoSpeciesHGT,[0,max(timespan)],initialHGT,method='RK45',t_eval=timespan)
                y2=np.transpose(sol.y)
                S1AbunHGT[i,k]=y2[-1,0]
                S2AbunHGT[i,k]=y2[-1,1]
            if max(S1AbunHGT[i,:])-min(S1AbunHGT[i,:])>thresh and max(S2AbunHGT[i,:])-min(S2AbunHGT[i,:])>thresh:
                Indd[i]=1
        feasi[sdf,wer]=sum(Indd)/rep
```

```python
plt.figure(figsize=(2,2))
for i in range(len(gammas)):
    plt.plot(etas,feasi[:,i],'o-',markersize=7,fillstyle='none')
plt.xlabel('transfer rate eta',fontsize=20)
plt.ylabel('bistability chance',fontsize=20)
plt.ylim(0,0.6)
plt.tick_params(direction='in',top='true',right='true')
plt.savefig('feasibility.pdf')
plt.show()

import pandas as pd
df=pd.DataFrame(feasi)
df.to_csv("feasi.csv")
```
